```python
food = pd.read_csv("recipes.csv")
```

```python
food = food.drop(columns = ['RecipeId',  'AuthorId', 'AuthorName', 
       'DatePublished', 'Description', 'Images', 'RecipeCategory',
       'ReviewCount', 'Calories', 'FatContent',
       'SaturatedFatContent', 'CholesterolContent', 'SodiumContent',
       'CarbohydrateContent', 'FiberContent', 'SugarContent', 'ProteinContent',
       'RecipeServings', 'RecipeYield',"CookTime","PrepTime","TotalTime" ])
```

```python
food.head()
```

```python
food_chicken = food.loc[["Chicken" in i for i in food["Name"]]]
food_chicken = food_chicken.drop(columns = "AggregatedRating")
food_chicken.head()
```

```python
new_cols = ["Name","RecipeIngredientQuantities","RecipeIngredientParts","RecipeInstructions", "Keywords"]
food_chicken=food_chicken.reindex(columns=new_cols)
food_chicken.head()
```

```python
food_chicken.to_csv("chicken_food.csv", index = False)
```

```python
chicken.rename(columns = {'title':'Name', 'NER':'Keywords'}, inplace = True)
chicken["extra"] = chicken["Name"]
chicken
```

```python
from sqlalchemy import create_engine
import sqlite3
engine = create_engine('sqlite://', echo=False)
from sqlalchemy import create_engine, inspect, func
```

```python
chicken.to_sql('nlg', con = engine)
food_chicken.to_sql('food', con=engine)
```
